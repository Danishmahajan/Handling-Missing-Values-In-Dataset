# Adding Missing Category

###### Suppose if you have more frequent categories, we just replace NAN with a new category

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('loan.csv',usecols=['BsmtQual','FireplaceQu', 'GarageType', 'SalePrice'])
df
```

```python
df.isnull().sum()
```

####  Plots Before Adding the Missing category in NAN values

```python
df['BsmtQual'].value_counts().plot.bar(color='orchid')
plt.title("Before Adding Missing in BsmtQual ")
```

```python
df['FireplaceQu'].value_counts().plot.bar(color='lightseagreen')
plt.title("Before Adding Missing in FireplaceQu ")
```

```python
df['GarageType'].value_counts().plot.bar(color='firebrick')
plt.title("Before Adding Missing in Garagetype ")
```

```python
def impute_nan(df,var):
    df[var+'_Missing']=np.where(df[var].isnull(),'Missing',df[var]) # or You can make change in original feature
    
```

```python
for feature in ['BsmtQual','FireplaceQu', 'GarageType']:
    impute_nan(df,feature)
```

```python
df.head(50)
```

```python
df.drop(['BsmtQual','FireplaceQu', 'GarageType'],axis=1,inplace=True)
```

####  Plots After Adding the Missing category in NAN values

```python
df['BsmtQual_Missing'].value_counts().plot.bar(color='orchid')
plt.title("After Adding Missing in BsmtQual ")
```

```python
df['FireplaceQu_Missing'].value_counts().plot.bar(color='lightseagreen')
plt.title("After Adding Missing in FireplaceQu ")
```

```python
df['GarageType_Missing'].value_counts().plot.bar(color='firebrick')
plt.title("After Adding Missing in Garagetype ")
```
